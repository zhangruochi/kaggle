```python
kf = KFold(n_splits=10)

clf = LogisticRegression(solver = "lbfgs",C = 1,tol=1e-4, max_iter = 500)


#train_features = all_features.loc[:,selected_features]
train_features = all_features.values


estimator = OneVsOneClassifier(clf)
scores = []
for train_index, test_index in kf.split(train_features):
    X_train, X_test = train_features[train_index], train_features[test_index]
    y_train, y_test = labels[train_index], labels[test_index]
    estimator.fit(X_train,y_train)
    score = estimator.score(X_test,y_test)
    scores.append(score)
print(np.mean(scores))
```

## 参数调节

```python
def tune_parameters(training_dataset,labels):
    
    tuned_parameters = {
    "estimator__solver": ["lbfgs","sag"],
    "estimator__C": [0.1, 1,10],
    "estimator__max_iter": [500,1000],
    "estimator__fit_intercept": [True,False],
    }

    scores = ['accuracy']

    # 将数据集分成训练集和测试集
    X_train, X_test, y_train, y_test = train_test_split(
    training_dataset, labels, test_size=0.3, random_state=0)
                        
    for score in scores:
        print("# Tuning hyper-parameters for %s\n" % score)

        clf = GridSearchCV(OneVsRestClassifier(LogisticRegression()), tuned_parameters, cv = 5,
                           scoring='%s' % score)

        clf.fit(X_train, y_train)

        print("Best parameters set found on development set:\n")
        #输出最优的模型参数
        print(clf.best_params_)
        
        y_true, y_pred = y_test, clf.predict(X_test)
        
        print(classification_report(y_true, y_pred))
        
        
    return clf

clf = tune_parameters(all_features,labels)
```
